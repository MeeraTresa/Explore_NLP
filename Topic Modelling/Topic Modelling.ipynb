# Naive Bayes Classification
Use Bayesian classification to classify reviews into positive and negative using both the unigram and bigram versions of the bag of words and tf-idf representations from 1 and 2 above (e.g., using sklearn's MultinomialNB or nltk's NaiveBayesClassifier)

We represent a text document as a “bag of words” - only the frequency of appearance of each word matters. The predicted class is the one with the highest (posterior) probability among the set of classes C, given the document.Naive Bayes for Documents and Classes.png


We assume conditional independence and arrive at cNB, the class that Naive Bayes
(NB) predicts for a document d.


```python
positive_y = [1] * len(positive_texts)
negative_y = [0] * len(negative_texts)
```

```python
all_texts = positive_texts + negative_texts
all_y = positive_y + negative_y
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(all_texts, all_y, test_size=0.2, random_state=1)
```

```python
# Initialize CountVectorizer and TfidfVectorizer for unigrams and bigrams using the text in train data and test data
unigram_vectorizer = CountVectorizer()
X_train_unigram = unigram_vectorizer.fit_transform(X_train)
X_test_unigram = unigram_vectorizer.transform(X_test)

bigram_vectorizer = CountVectorizer(ngram_range=(2, 2))
X_train_bigram = bigram_vectorizer.fit_transform(X_train)
X_test_bigram = bigram_vectorizer.transform(X_test)

tfidf_unigram_vectorizer = TfidfVectorizer()
X_train_tfidf_unigram = tfidf_unigram_vectorizer.fit_transform(X_train)
X_test_tfidf_unigram = tfidf_unigram_vectorizer.transform(X_test)

tfidf_bigram_vectorizer = TfidfVectorizer(ngram_range=(2, 2))
X_train_tfidf_bigram = tfidf_bigram_vectorizer.fit_transform(X_train)
X_test_tfidf_bigram = tfidf_bigram_vectorizer.transform(X_test)
```

Naive Bayes classifier for multinomial models.

The multinomial Naive Bayes classifier is suitable for classification with discrete features (e.g., word counts for text classification). The multinomial distribution normally requires integer feature counts. However, in practice, fractional counts such as tf-idf may also work.

fit(X, y, sample_weight=None)
X {array-like, sparse matrix} of shape (n_samples, n_features)
Training vectors, where n_samples is the number of samples and n_features is the number of features.

y array-like of shape (n_samples,)
Target values.

```python
from sklearn.naive_bayes import MultinomialNB
```

Training and Predictions with the Multinomial NB using the term document matrix and tfidf

```python
nb_unigram = MultinomialNB()
nb_unigram.fit(X_train_unigram, y_train)

nb_bigram = MultinomialNB()
nb_bigram.fit(X_train_bigram, y_train)

nb_tfidf_unigram = MultinomialNB()
nb_tfidf_unigram.fit(X_train_tfidf_unigram, y_train)

nb_tfidf_bigram = MultinomialNB()
nb_tfidf_bigram.fit(X_train_tfidf_bigram, y_train)
# Predictions
y_pred_unigram = nb_unigram.predict(X_test_unigram)
y_pred_bigram = nb_bigram.predict(X_test_bigram)
y_pred_tfidf_unigram = nb_tfidf_unigram.predict(X_test_tfidf_unigram)
y_pred_tfidf_bigram = nb_tfidf_bigram.predict(X_test_tfidf_bigram)
```

Evaluation

```python
from sklearn.metrics import classification_report
from sklearn.metrics import precision_score
```

```python
print("Unigram CountVectorizer Classifier Report:\n", classification_report(y_test, y_pred_unigram))
print("Bigram CountVectorizer Classifier Report:\n", classification_report(y_test, y_pred_bigram))
print("Unigram TF-IDF Classifier Report:\n", classification_report(y_test, y_pred_tfidf_unigram))
print("Bigram TF-IDF Classifier Report:\n", classification_report(y_test, y_pred_tfidf_bigram))
```

```python
precision_unigram = precision_score(y_test, y_pred_unigram)
precision_bigram = precision_score(y_test, y_pred_bigram)
precision_tfidf_unigram = precision_score(y_test, y_pred_tfidf_unigram)
precision_tfidf_bigram = precision_score(y_test, y_pred_tfidf_bigram)
print("Precision (Unigram CountVectorizer):", precision_unigram)
print("Precision (Bigram CountVectorizer):", precision_bigram)
print("Precision (Unigram TF-IDF):", precision_tfidf_unigram)
print("Precision (Bigram TF-IDF):", precision_tfidf_bigram)
```


# Pointwise Mutual Information
Use nltk.collocations functions to calculate the top 10 bigram collocations based on PPMI


Pointwise Mutual Information (PMI) measures how often two events, x and y, co-occur relative
to how often they would co-occur by chance (i.e., if they were independent)
PMI.png

Collocations are expressions of multiple words which commonly co-occur.

https://www.nltk.org/howto/collocations.html

```python
from nltk.collocations import BigramAssocMeasures, BigramCollocationFinder
```

```python
# Create a BigramCollocationFinder with the bigram association measures
finder = BigramCollocationFinder.from_words(all_texts)
```

```python
pmi = BigramAssocMeasures.pmi
```

```python
# Find the top 10 bigram collocations based on PMI scores
top_10_collocations = finder.nbest(pmi, 10)
# Print the top 10 bigram collocations
for collocation in top_10_collocations:
    print(" ".join(collocation))
    print("-----------------------------------------------------------")
```

## Using sklearn models

```python
from __future__ import print_function
import pyLDAvis
import pyLDAvis.lda_model
pyLDAvis.enable_notebook()
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
categories = ['sci.med', 'sci.space', 'talk.politics.guns']
newsgroups = fetch_20newsgroups(subset='all',categories=categories,remove=('headers', 'footers', 'quotes'))
docs_raw = newsgroups.data
print(len(docs_raw))
```

```python
from pprint import pprint
pprint(list(newsgroups.target_names))
```

```python
# Convert to document-term matrix
# Next, the raw documents are converted into document-term matrix, possibly as raw counts or in TF-IDF form.
tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                                stop_words = 'english',
                                lowercase = True,
                                token_pattern = r'\b[a-zA-Z]{3,}\b',
                                max_df = 0.5,
                                min_df = 10)
dtm_tf = tf_vectorizer.fit_transform(docs_raw)
print(dtm_tf.shape)
```

```python
tfidf_vectorizer = TfidfVectorizer(**tf_vectorizer.get_params())
dtm_tfidf = tfidf_vectorizer.fit_transform(docs_raw)
print(dtm_tfidf.shape)
```

```python
# Fit Latent Dirichlet Allocation models - this step takes a while!

# for TF DTM
lda_tf = LatentDirichletAllocation(n_components=7, random_state=0)
lda_tf.fit(dtm_tf)
# for TFIDF DTM
lda_tfidf = LatentDirichletAllocation(n_components=7, random_state=0)
lda_tfidf.fit(dtm_tfidf)
```

```python
# Visualizing the models with pyLDAvis, using the term frequency (tf) vectors
pyLDAvis.lda_model.prepare(lda_tf, dtm_tf, tf_vectorizer)
```

```python
# Visualizing the models with pyLDAvis, using the tf-idf vectors

pyLDAvis.lda_model.prepare(lda_tfidf, dtm_tfidf, tfidf_vectorizer)
```

## Using genesim Models
https://radimrehurek.com/gensim_3.8.3/index.html

```python
import pyLDAvis.gensim_models


import gensim
from gensim import corpora
from gensim.parsing.preprocessing import remove_stopwords
from gensim.parsing.preprocessing import strip_punctuation
from gensim.parsing.preprocessing import strip_short


from sklearn.datasets import fetch_20newsgroups

from IPython.core.display import HTML
```

```python
# Load the 20newsgroups dataset
categories = ['sci.med', 'sci.space', 'talk.politics.guns']
newsgroups = fetch_20newsgroups(subset='all', categories=categories, shuffle=True, remove=('headers', 'footers', 'quotes'))
docs_raw = newsgroups.data
print(len(docs_raw))
```

```python
docs_raw[0]
```

```python
def remove_stopwords_from_string(text):
    return remove_stopwords(strip_short(strip_punctuation(text.lower()), minsize=2))
```

```python
docs_raw = [remove_stopwords_from_string(doc) for doc in docs_raw]
print(len(docs_raw))
```

```python
docs_raw[0]
```

```python
# Define the document and the corpus for the data
id2word = corpora.Dictionary([text.lower().split() for text in docs_raw])
corpus = [id2word.doc2bow(text.lower().split()) for text in docs_raw]
```

```python
len(corpus)
```

```python
[[(id2word[i], freq) for i, freq in doc] for doc in corpus[:1]]
```

```python
# Define different numbers of topics
num_topics_list = [4, 7, 10]
pyLDAvis.enable_notebook()
# Create LDA models and visualize topics
for num_topics in num_topics_list:
  lda_model = gensim.models.LdaModel(corpus, id2word=id2word, num_topics=num_topics, random_state=0)
  lda_display = pyLDAvis.gensim_models.prepare(lda_model, corpus, id2word, sort_topics=False)
  pyLDAvis.save_html(lda_display, 'lda_viz.html')
  display(HTML('lda_viz.html'))
```
