```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
sentence = "Workers at Target stores and distribution centers in places like New York, where competition for finding and hiring staff is the fiercest, could see starting  wages as high as $24 an hour this year"
```

```python
doc = nlp(sentence)
```

```python
for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

```python
for ent in doc.ents:
  print(spacy.explain(ent.label_))
```

```python
from spacy import displacy
```

```python
displacy.render(doc, style="ent",jupyter=True)
```

```python
displacy.render(doc, style="dep",jupyter=True)
```
